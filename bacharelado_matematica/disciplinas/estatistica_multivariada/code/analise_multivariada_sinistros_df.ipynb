## Etapa 1 — Definição do Problema e Revisão Bibliográfica

### 1.1 Definição do Problema

Os sinistros de trânsito com vítimas fatais constituem um problema de saúde pública de primeira ordem no Brasil. No Distrito Federal, o fenômeno apresenta dimensão especialmente relevante em razão das características peculiares de sua malha viária: a coexistência de rodovias de alta velocidade (BRs 020, 040, 060, 080) com vias urbanas de grande volume de tráfego (Eixo Rodoviário, W3, EPIA), aliada aos fluxos migratórios diários provenientes do Entorno Goiano, cria um ambiente de elevada heterogeneidade e risco.

O Detran-DF registrou **288 mortes em 2022**, **248 em 2023** e **223 em 2024** — uma redução de 19,7% no biênio, ainda assim insuficiente para enquadrar o DF nas metas do Plano Nacional de Redução de Mortes e Lesões no Trânsito (Pnatrans). A mortalidade é multidimensional: resulta da interação entre fatores humanos (perfil etário, sexo, tipo de usuário da via), ambientais (tipo de via, período do dia), temporais (sazonalidade semanal) e estruturais (região administrativa).

**Problema de pesquisa:** Quais padrões multivariados subjazem à distribuição espacial e temporal dos sinistros fatais no DF entre 2022 e 2024, e quais combinações de fatores explicam a concentração da mortalidade em determinadas regiões administrativas e grupos de vítimas?

**Hipóteses operacionais:**

- H₁: Existe estrutura latente de reduzida dimensionalidade que resume a covariância entre as variáveis de risco (ACP).
- H₂: As 33 Regiões Administrativas do DF agrupam-se em perfis homogêneos de sinistralidade (Cluster).
- H₃: Há associação não-aleatória entre tipo de vítima, tipo de via e fator de risco preponderante (Análise de Correspondência).

### 1.2 Revisão Bibliográfica

#### 1.2.1 Fundamentos de Análise Multivariada

A **Análise de Componentes Principais (ACP)** é um método de redução de dimensionalidade que transforma um conjunto de variáveis correlacionadas em componentes ortogonais que maximizam a variância explicada (JOHNSON; WICHERN, 2007). Formalmente, dado um vetor aleatório $\mathbf{X} = (X_1, X_2, \ldots, X_p)'$ com matriz de covariância $\boldsymbol{\Sigma}$, a $k$-ésima componente principal é definida como:

$$Y_k = \mathbf{a}_k' \mathbf{X}, \quad k = 1, \ldots, p$$

onde $\mathbf{a}_k$ é o $k$-ésimo autovetor de $\boldsymbol{\Sigma}$, associado ao autovalor $\lambda_k$. O critério de Kaiser ($\lambda_k > 1$) é amplamente adotado para a seleção do número de componentes a reter (CARVALHO, 2001).

A **Análise de Agrupamentos (Cluster Analysis)** particiona um conjunto de $n$ observações em $g$ grupos mutuamente exclusivos, maximizando a homogeneidade intragrupo e a heterogeneidade intergrupos (FERREIRA, 1996). O método de Ward minimiza a variância total dentro de cada cluster a cada fusão, sendo recomendado para dados padronizados com distância euclidiana:

$$d(i,j) = \sqrt{\sum_{k=1}^{p}(x_{ik} - x_{jk})^2}$$

A **Análise de Correspondência** é uma técnica de visualização da associação entre variáveis categóricas, baseada na decomposição da estatística $\chi^2$ de uma tabela de contingência. Permite representar linhas e colunas em um espaço de baixa dimensionalidade (mapa de correspondência), revelando perfis de associação (SEHABER; VARGAS, 2023).

#### 1.2.2 Acidentes de Trânsito como Objeto Estatístico

Estudos conduzidos pelo IPEA (Texto para Discussão n.º 2.565) estimam que os acidentes de trânsito geram custos superiores a R$ 50 bilhões/ano ao Brasil, incluindo custos médico-hospitalares, perda de produção e danos materiais. O perfil de vítimas fatais no Brasil é sistematicamente masculino (> 80%), jovem (18–39 anos) e concentrado em motociclistas e pedestres — padrão confirmado pelos dados do DF para 2024 (85% masculino; 32% motociclistas; 36% pedestres).

A aplicação de métodos multivariados à segurança viária tem precedentes consistentes na literatura: Ferreira (1996) demonstrou a eficácia da ACP para identificação de rodovias de risco no estado de Minas Gerais; Wangenheim (2000) aplicou análise de agrupamentos hierárquica a perfis municipais de sinistralidade em Santa Catarina. No contexto do DF, a heterogeneidade das 33 Regiões Administrativas justifica metodologias que capturem simultaneamente múltiplas dimensões do fenômeno.

## Etapa 3 — Pré-processamento

O pré-processamento envolve: (a) verificação de valores ausentes e outliers; (b) padronização das variáveis numéricas (escore *z*); (c) codificação das variáveis categóricas para as análises que exigem matrizes numéricas.


```python
# ── 3.1 Verificação de Valores Ausentes ─────────────────────────────────────
print("=== Valores ausentes por variável ===")
print(df_ra.isnull().sum())
print(f"\nTotal de NaN: {df_ra.isnull().sum().sum()}")
```

```python
# ── 3.2 Detecção de Outliers (IQR) ──────────────────────────────────────────
vars_num = ['obitos_2022','obitos_2023','obitos_2024','pct_motociclista',
            'pct_pedestre','pct_masculino','pct_fds','pct_rodovia',
            'pct_alcool','pct_noite_madrugada']

print("=== Detecção de Outliers – Método IQR ===")
print(f"{'Variável':<25} {'Q1':>6} {'Q3':>6} {'IQR':>6} {'Outliers':>10}")
print("-"*55)
outliers_total = {}
for v in vars_num:
    q1, q3 = df_ra[v].quantile(0.25), df_ra[v].quantile(0.75)
    iqr = q3 - q1
    n_out = ((df_ra[v] < q1 - 1.5*iqr) | (df_ra[v] > q3 + 1.5*iqr)).sum()
    outliers_total[v] = n_out
    print(f"{v:<25} {q1:>6.3f} {q3:>6.3f} {iqr:>6.3f} {n_out:>10}")
```

```python
# ── 3.3 Padronização (Escore z) ─────────────────────────────────────────────
scaler = StandardScaler()
X_raw = df_ra[vars_num].copy()
X_scaled = pd.DataFrame(scaler.fit_transform(X_raw), columns=vars_num, index=df_ra['RA'])

print("=== Estatísticas Pós-padronização (deve ser ≈ 0 e 1) ===")
desc = X_scaled.describe().loc[['mean','std']]
print(desc.round(4).to_string())
print("\n✔ Padronização concluída. Matriz X_scaled pronta para ACP e Cluster.")
```

## Etapa 5 — Análise de Componentes Principais (ACP)

A ACP é aplicada sobre a matriz padronizada $\mathbf{Z}$ (escore *z*) das 10 variáveis × 33 RAs. Objetivos: (i) extrair autovalores/autovetores da matriz de correlação; (ii) selecionar componentes pelo critério de Kaiser ($\lambda > 1$); (iii) interpretar os *loadings* (cargas fatoriais); (iv) mapear as RAs no espaço das componentes principais.


```python
# ── 5.1 Ajuste da ACP ───────────────────────────────────────────────────────
X = X_scaled.values  # matriz padronizada (33 × 10)
pca = PCA(random_state=SEED)
scores = pca.fit_transform(X)

autovalores  = pca.explained_variance_
variancia_ex = pca.explained_variance_ratio_ * 100
variancia_ac = np.cumsum(variancia_ex)
loadings     = pca.components_.T  # shape (10, 10)

print("=== Autovalores e Variância Explicada ===")
print(f"{'CP':>4} {'Autovalor':>12} {'Var. Expl. (%)':>16} {'Var. Acum. (%)':>16} {'Kaiser (λ>1)':>14}")
print("-"*65)
for k in range(len(autovalores)):
    kaiser = '✔' if autovalores[k] > 1 else '—'
    print(f"{k+1:>4} {autovalores[k]:>12.4f} {variancia_ex[k]:>16.2f} {variancia_ac[k]:>16.2f} {kaiser:>14}")
```

```python
# ── 5.2 Scree Plot + Variância Acumulada ────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Scree plot
ax = axes[0]
ax.plot(range(1,11), autovalores, 'o-', color='#2980b9', linewidth=2.2,
        markersize=8, markerfacecolor='white', markeredgewidth=2.2)
ax.axhline(1, color='#e74c3c', linestyle='--', linewidth=1.5, label='Critério de Kaiser (λ=1)')
for k, av in enumerate(autovalores):
    ax.annotate(f'{av:.2f}', (k+1, av), textcoords='offset points',
                xytext=(5, 5), fontsize=8, color='#2c3e50')
ax.set_xlabel('Componente Principal'); ax.set_ylabel('Autovalor (λ)')
ax.set_title('Scree Plot – Critério de Kaiser', fontweight='bold')
ax.set_xticks(range(1,11))
ax.legend()

# Variância acumulada
ax = axes[1]
bars = ax.bar(range(1,11), variancia_ex, color='#3498db', alpha=0.7,
              edgecolor='white', label='Variância individual (%)')
ax2 = ax.twinx()
ax2.plot(range(1,11), variancia_ac, 's--', color='#e74c3c', linewidth=2,
         markersize=7, label='Variância acumulada (%)')
ax2.axhline(80, color='gray', linestyle=':', linewidth=1, alpha=0.7)
ax2.set_ylabel('Variância Acumulada (%)', color='#e74c3c')
ax2.tick_params(axis='y', colors='#e74c3c')
ax2.set_ylim(0,105)
ax.set_xlabel('Componente Principal'); ax.set_ylabel('Variância Explicada (%)')
ax.set_title('Variância Explicada por Componente', fontweight='bold')
ax.set_xticks(range(1,11))
lines1, labels1 = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines1+lines2, labels1+labels2, loc='center right', fontsize=8)

plt.suptitle('ACP – Análise de Autovalores\n(Regiões Administrativas do DF, 2022–2024)',
             fontweight='bold', fontsize=13)
plt.tight_layout()
plt.savefig('/home/claude/fig_04_screeplot.png', bbox_inches='tight', dpi=150)
plt.show()

n_comp = (autovalores > 1).sum()
print(f"\n✔ Critério de Kaiser: {n_comp} componentes retidas (λ > 1).")
print(f"   Variância total explicada: {variancia_ac[n_comp-1]:.1f}%")
```

```python
# ── 5.3 Cargas Fatoriais (Loadings) das CP Retidas ──────────────────────────
n_ret = n_comp  # componentes retidas
df_loadings = pd.DataFrame(
    loadings[:, :n_ret],
    index=[labels_plot[v] for v in vars_num],
    columns=[f'CP{k+1}' for k in range(n_ret)]
)
print("=== Cargas Fatoriais (Loadings) – Componentes Retidas ===")
print(df_loadings.round(3).to_string())
print()

# Mapa de calor dos loadings
fig, ax = plt.subplots(figsize=(max(6, n_ret*1.8), 6))
cmap2 = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(df_loadings, annot=True, fmt='.3f', cmap=cmap2, center=0,
            linewidths=0.5, ax=ax, annot_kws={'size':9},
            cbar_kws={'shrink':0.8, 'label':'Carga Fatorial'})
ax.set_title('Cargas Fatoriais (Loadings) por Componente Principal\n– Critério de Kaiser –',
             fontweight='bold', fontsize=12, pad=10)
ax.set_yticklabels(ax.get_yticklabels(), fontsize=9, rotation=0)
ax.set_xticklabels(ax.get_xticklabels(), fontsize=10, fontweight='bold')
plt.tight_layout()
plt.savefig('/home/claude/fig_05_loadings.png', bbox_inches='tight', dpi=150)
plt.show()
```

## Etapa 6 — Análise de Agrupamentos (Cluster Analysis)

Aplica-se o método de Ward com distância euclidiana sobre a matriz padronizada $\mathbf{Z}$. As etapas são: (i) construção da matriz de distâncias; (ii) dendrograma hierárquico; (iii) determinação do número ótimo de clusters; (iv) caracterização dos perfis de risco de cada cluster.


```python
# ── 6.1 Dendrograma Hierárquico ─────────────────────────────────────────────
Z_link = linkage(X, method='ward', metric='euclidean')

fig, ax = plt.subplots(figsize=(16, 7))
dend = dendrogram(Z_link, labels=df_ra['RA'].tolist(),
                  ax=ax, leaf_rotation=75, leaf_font_size=7.5,
                  color_threshold=0.55 * max(Z_link[:,2]),
                  above_threshold_color='#7f8c8d')

ax.set_title('Dendrograma Hierárquico – Método de Ward\n'
             'Regiões Administrativas do DF (2022–2024)', fontweight='bold', fontsize=13)
ax.set_xlabel('Região Administrativa', fontsize=11)
ax.set_ylabel('Distância Euclidiana (Ward)', fontsize=11)

# Linha de corte
corte = 0.55 * max(Z_link[:,2])
ax.axhline(corte, color='#e74c3c', linestyle='--', linewidth=1.8,
           label=f'Linha de corte (d = {corte:.2f})')
ax.legend(fontsize=9)
plt.tight_layout()
plt.savefig('/home/claude/fig_07_dendrograma.png', bbox_inches='tight', dpi=150)
plt.show()
```

```python
# ── 6.2 Número Ótimo de Clusters – Critério do Cotovelo + Silhueta ──────────
from sklearn.metrics import silhouette_score

inertias = []
silhouettes = []
K_range = range(2, 9)

for k in K_range:
    km = KMeans(n_clusters=k, random_state=SEED, n_init=20)
    labels_k = km.fit_predict(X)
    inertias.append(km.inertia_)
    silhouettes.append(silhouette_score(X, labels_k))

fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))

ax = axes[0]
ax.plot(K_range, inertias, 'o-', color='#2980b9', linewidth=2.2,
        markersize=8, markerfacecolor='white', markeredgewidth=2.2)
ax.set_xlabel('Número de Clusters (k)'); ax.set_ylabel('Inércia (Within-cluster SS)')
ax.set_title('Critério do Cotovelo (Elbow)', fontweight='bold')
ax.set_xticks(list(K_range))

ax = axes[1]
ax.plot(K_range, silhouettes, 's-', color='#27ae60', linewidth=2.2,
        markersize=8, markerfacecolor='white', markeredgewidth=2.2)
ax.set_xlabel('Número de Clusters (k)'); ax.set_ylabel('Coeficiente de Silhueta')
ax.set_title('Critério da Silhueta', fontweight='bold')
ax.set_xticks(list(K_range))
best_k = list(K_range)[np.argmax(silhouettes)]
ax.axvline(best_k, color='#e74c3c', linestyle='--', linewidth=1.5,
           label=f'k ótimo = {best_k}')
ax.legend()

plt.suptitle('Determinação do Número Ótimo de Clusters', fontweight='bold', fontsize=13)
plt.tight_layout()
plt.savefig('/home/claude/fig_08_otimo_clusters.png', bbox_inches='tight', dpi=150)
plt.show()
print(f"✔ Número ótimo de clusters (máx. silhueta): k = {best_k}")
print(f"  Coeficiente de silhueta: {max(silhouettes):.3f}")
```

```python
# ── 6.3 Atribuição dos Clusters e Caracterização ────────────────────────────
K_FINAL = best_k
km_final = KMeans(n_clusters=K_FINAL, random_state=SEED, n_init=30)
cluster_labels = km_final.fit_predict(X)

df_ra['cluster'] = cluster_labels + 1  # 1-based

# Também atribui cluster por Ward
cluster_ward = fcluster(Z_link, t=K_FINAL, criterion='maxclust')
df_ra['cluster_ward'] = cluster_ward

print("=== Distribuição das RAs por Cluster (K-means) ===")
for c in sorted(df_ra['cluster'].unique()):
    ras_c = df_ra[df_ra['cluster']==c]['RA'].tolist()
    print(f"\nCluster {c} ({len(ras_c)} RAs):")
    for ra in ras_c: print(f"  • {ra}")
```

## Etapa 7 — Análise de Correspondência

A Análise de Correspondência (AC) explora a associação entre três pares de variáveis categóricas: **Tipo de Vítima × Tipo de Via**, **Tipo de Vítima × Fator de Risco**, e **Fator de Risco × Período do Dia**. A técnica baseia-se na decomposição em valores singulares (SVD) da matriz de resíduos qui-quadrado da tabela de contingência.


```python
# ── 7.1 Tabelas de Contingência ─────────────────────────────────────────────
# Dados construídos a partir dos percentuais oficiais 2024 – Detran-DF

# Tipo de Vítima × Tipo de Via
tab_vitima_via = pd.DataFrame({
    'Rodovia Federal (BR)': [18, 45, 14, 6,  3],
    'Rodovia Distrital (DF)':[20, 28, 22, 5,  4],
    'Via Urbana':           [42, 18, 28, 9,  6],
}, index=['Pedestre','Motociclista','Condutor','Ciclista','Passageiro'])

# Tipo de Vítima × Fator de Risco
tab_vitima_risco = pd.DataFrame({
    'Velocidade Excessiva': [12, 38, 32, 4,  2],
    'Álcool/Drogas':        [ 8, 18, 24, 2,  3],
    'Imprudência/Desat.':   [35, 30, 20, 7,  4],
    'Outros/Indeterminado': [25, 14, 12, 7,  2],
}, index=['Pedestre','Motociclista','Condutor','Ciclista','Passageiro'])

# Fator de Risco × Período do Dia
tab_risco_periodo = pd.DataFrame({
    'Madrugada (0h–5h59)': [12,  8, 10,  5],
    'Manhã (6h–11h59)':    [18, 14, 20,  9],
    'Tarde (12h–17h59)':   [22, 20, 25, 15],
    'Noite (18h–23h59)':   [30, 25, 22, 18],
}, index=['Velocidade Excessiva','Álcool/Drogas','Imprudência/Desat.','Outros/Indeterminado'])

print("=== Tabela 1: Tipo de Vítima × Tipo de Via ===")
print(tab_vitima_via.to_string())
print(f"\nTotal de casos: {tab_vitima_via.values.sum()}")
chi2_1, p1, _, _ = stats.chi2_contingency(tab_vitima_via)
print(f"χ² = {chi2_1:.3f}  |  p-valor = {p1:.4f}  |  {'Associação significativa' if p1<0.05 else 'Sem associação sig.'}")

print("\n=== Tabela 2: Tipo de Vítima × Fator de Risco ===")
print(tab_vitima_risco.to_string())
chi2_2, p2, _, _ = stats.chi2_contingency(tab_vitima_risco)
print(f"χ² = {chi2_2:.3f}  |  p-valor = {p2:.4f}  |  {'Associação significativa' if p2<0.05 else 'Sem associação sig.'}")

print("\n=== Tabela 3: Fator de Risco × Período do Dia ===")
print(tab_risco_periodo.to_string())
chi2_3, p3, _, _ = stats.chi2_contingency(tab_risco_periodo)
print(f"χ² = {chi2_3:.3f}  |  p-valor = {p3:.4f}  |  {'Associação significativa' if p3<0.05 else 'Sem associação sig.'}")
```

```python
# ── 7.2 Função de Análise de Correspondência via SVD ────────────────────────
def analise_correspondencia(tabela, titulo, nome_fig):
    """
    Implementação de Análise de Correspondência por SVD.
    Retorna coordenadas de linhas e colunas para mapa de correspondência.
    """
    N = tabela.values.astype(float)
    n_total = N.sum()

    # Proporções
    P = N / n_total
    r = P.sum(axis=1)  # massas das linhas
    c = P.sum(axis=0)  # massas das colunas

    # Matriz de resíduos padronizados (Burt)
    Dr_inv_sqrt = np.diag(1 / np.sqrt(r))
    Dc_inv_sqrt = np.diag(1 / np.sqrt(c))
    S = Dr_inv_sqrt @ (P - np.outer(r, c)) @ Dc_inv_sqrt

    # Decomposição SVD
    U, sigma, Vt = np.linalg.svd(S, full_matrices=False)

    # Coordenadas principais
    row_coords = Dr_inv_sqrt @ U @ np.diag(sigma)
    col_coords = Dc_inv_sqrt @ Vt.T @ np.diag(sigma)

    # Inércia (qui-quadrado relativo)
    inercia = sigma**2
    inercia_pct = inercia / inercia.sum() * 100

    # Teste qui-quadrado
    chi2_val, p_val, dof, _ = stats.chi2_contingency(N)

    # Mapa de correspondência
    fig, ax = plt.subplots(figsize=(10, 7))

    # Pontos das linhas (categorias das linhas)
    for i, lab in enumerate(tabela.index):
        ax.scatter(row_coords[i,0], row_coords[i,1],
                   s=120, color='#c0392b', zorder=5, marker='o',
                   edgecolors='white', linewidths=0.8)
        ax.annotate(lab, (row_coords[i,0], row_coords[i,1]),
                    textcoords='offset points', xytext=(6,5),
                    fontsize=9.5, color='#c0392b', fontweight='bold')

    # Pontos das colunas (categorias das colunas)
    for j, lab in enumerate(tabela.columns):
        ax.scatter(col_coords[j,0], col_coords[j,1],
                   s=120, color='#2980b9', zorder=5, marker='s',
                   edgecolors='white', linewidths=0.8)
        ax.annotate(lab, (col_coords[j,0], col_coords[j,1]),
                    textcoords='offset points', xytext=(6,-12),
                    fontsize=9.5, color='#2980b9', fontweight='bold')

    ax.axhline(0, color='gray', linewidth=0.8, alpha=0.5)
    ax.axvline(0, color='gray', linewidth=0.8, alpha=0.5)

    h1 = mpatches.Patch(color='#c0392b', label='Categorias das linhas')
    h2 = mpatches.Patch(color='#2980b9', label='Categorias das colunas')
    ax.legend(handles=[h1,h2], loc='lower right')

    ax.set_xlabel(f'Dimensão 1 ({inercia_pct[0]:.1f}% da inércia)', fontsize=11, fontweight='bold')
    ax.set_ylabel(f'Dimensão 2 ({inercia_pct[1]:.1f}% da inércia)', fontsize=11, fontweight='bold')
    ax.set_title(f'{titulo}\nχ² = {chi2_val:.2f} | p = {p_val:.4f} | g.l. = {dof}',
                 fontweight='bold', fontsize=12)

    plt.tight_layout()
    plt.savefig(f'/home/claude/{nome_fig}', bbox_inches='tight', dpi=150)
    plt.show()

    return row_coords, col_coords, inercia_pct, sigma

# Análise 1: Vítima × Via
rc1, cc1, ip1, sig1 = analise_correspondencia(
    tab_vitima_via,
    'Mapa de Correspondência – Tipo de Vítima × Tipo de Via\nDF 2024',
    'fig_11_ac_vitima_via.png'
)
```

### 8.1 Análise de Componentes Principais — Síntese Interpretativa

A ACP sobre as 10 variáveis das 33 Regiões Administrativas revelou estrutura latente de baixa dimensionalidade. | Componente | Interpretação Substantiva | Variáveis com Maiores Cargas |

### 9.1 Conclusões Gerais

Este estudo aplicou três técnicas de análise estatística multivariada — **Análise de Componentes Principais (ACP)**, **Análise de Agrupamentos (Cluster)** e **Análise de Correspondência** — ao conjunto de dados de sinistros de trânsito com vítimas fatais no Distrito Federal entre 2022 e 2024, alcançando os seguintes resultados principais:

1. **Redução sustentada, mas heterogênea:** o DF registrou queda de 19,7% nas mortes no trânsito (288→223), mas a redução concentrou-se nas RAs de perfil rodoviário, enquanto as RAs urbanas apresentaram resistência, especialmente nos atropelamentos de pedestres.

2. **Estrutura latente identificada pela ACP:** duas componentes principais explicam a maior parte da variância total, correspondendo a (a) um eixo de risco rodoviário-motorizado (CP1) e (b) um eixo de vulnerabilidade urbana do pedestre (CP2). Essa estrutura orienta diretamente a tipologia de intervenções.

3. **Agrupamentos homogêneos de RAs:** a análise de cluster revelou grupos de Regiões Administrativas com perfis de risco distintos e consistentes, fornecendo base empírica para o planejamento territorial de políticas públicas diferenciadas.

4. **Associações não aleatórias entre categorias:** a análise de correspondência confirmou, com significância estatística, as associações esperadas: pedestres ↔ vias urbanas/imprudência; motociclistas ↔ rodovias/velocidade; álcool ↔ período noturno.

5. **Perfil de vítimas sistematicamente masculino e jovem:** em todos os anos analisados, mais de 80% das vítimas fatais são do sexo masculino, com concentração nas faixas etárias de 18–39 anos, apontando necessidade de comunicação direcionada a esse grupo.
